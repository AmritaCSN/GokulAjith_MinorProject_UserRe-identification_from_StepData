```python
# Count the occurrences of each user in the 'user' column
user_counts = data['user'].value_counts()

# Display the count of each user
print(user_counts.sort_index())
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame and 'user' is the target column
df_sorted = data.sort_values(by='user').reset_index(drop=True)
df_sorted_val = data_val.sort_values(by='user').reset_index(drop=True)
```

```python
# Get unique users
users = df_sorted['user'].unique()

# Dictionary to hold data for each user
user_data = {user: df_sorted[df_sorted['user'] == user] for user in users}


users_val = df_sorted['user'].unique()

# Dictionary to hold data for each user
user_data_val = {user: df_sorted_val[df_sorted_val['user'] == user] for user in users_val}
```

```python
import numpy as np

sequence_length = 25# Number of days in each sequence
X = []
y = []

for user, data in user_data.items():
    user_features = data.drop('user', axis=1).values  # Features
    user_labels = data['user'].values  # Labels (all same for each user)

    # Number of possible sequences for this user
    num_sequences = len(data) - sequence_length

    for i in range(num_sequences):
        # Extract sequence of features
        seq_x = user_features[i:i + sequence_length]
        X.append(seq_x)
        
        # Label is the user of the last day in the sequence
        seq_y = user_labels[i + sequence_length]
        y.append(seq_y)

# Convert to numpy arrays
X = np.array(X)  # Shape: (total_sequences, sequence_length, n_features)
y = np.array(y)  # Shape: (total_sequences,)
```

```python
import numpy as np

sequence_length = 25# Number of days in each sequence
X_val = []
y_val = []

for user, data in user_data_val.items():
    
    user_features = data.drop('user', axis=1).values  # Features
    user_labels = data['user'].values  # Labels (all same for each user)

    # Number of possible sequences for this user
    num_sequences = len(data) - sequence_length

    for i in range(num_sequences):
        # Extract sequence of features
        seq_x = user_features[i:i + sequence_length]
        X_val.append(seq_x)
        
        # Label is the user of the last day in the sequence
        seq_y = user_labels[i + sequence_length]
        y_val.append(seq_y)

# Convert to numpy arrays
X_val = np.array(X_val)  # Shape: (total_sequences, sequence_length, n_features)
y_val = np.array(y_val)  # Shape: (total_sequences,)
```

```python
unique, counts = np.unique(y_val, return_counts=True)
label_counts = dict(zip(unique, counts))
label_counts
```

```python
import keras_tuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout,GRU,Bidirectional
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
def build_model(hp):
    model = Sequential()
    
    # Tune the number of LSTM units
    model.add(Bidirectional(LSTM(units=hp.Int('units', min_value=32, max_value=256, step=2), 
                   input_shape=(sequence_length, n_features))))
    
    # Tune dropout rate
    model.add(Dropout(rate=hp.Float('dropout_rate', min_value=0.1, max_value=0.5, step=0.1)))
    
    # Fixed Dense layer, not tuning it
    model.add(Dense(32, activation='softmax'))  
    
    # Tune the learning rate for the Adam optimizer
    model.compile(optimizer=Adam(learning_rate=hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])),
                  loss='sparse_categorical_crossentropy', 
                  metrics=['accuracy'])
    
    return model
# Define callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1) 
```

```python
from keras_tuner import RandomSearch
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
n_features = X.shape[2]
n_users = len(users)

tuner = RandomSearch(
    build_model,
    objective='val_loss',  # Optimize for validation accuracy
    max_trials=30,  # Number of different hyperparameter combinations to try
    executions_per_trial=2,  # Number of models to train for each combination
    directory='my_dir',
    project_name='lstm_tuning'
)

# Perform the hyperparameter search
tuner.search(X, y, epochs=100, batch_size=32, validation_data=(X_val, y_val),
             callbacks=[early_stopping, lr_scheduler])
```

```python
# Get the optimal hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

# print(f"Best number of LSTM units: {best_hps.get('units')}")
# print(f"Best dropout rate: {best_hps.get('dropout_rate')}")

# print(f"Best learning rate: {best_hps.get('learning_rate')}")
best_hps
```

```python
# Build and train the best model
best_model = tuner.hypermodel.build(best_hps)
history = best_model.fit(X, y, epochs=50, validation_data=(X_val, y_val),
                         callbacks=[early_stopping, lr_scheduler])
```
